```python
import numpy as np
import seaborn as sns
%matplotlib inline
dist = np.random.normal(loc=10, size=100000)
dist = dist[dist > 0]
sns.distplot(dist)
```

```python
dist1 = np.power(dist, 4)
sns.distplot(dist1)
```

```python
dist2 = np.power(dist, 1/6)
sns.distplot(dist2)
```

```python
from scipy.stats import boxcox_normmax
dist3 = np.power(2, dist)
print(boxcox_normmax(dist3))
sns.distplot(dist3)
```
